```python
import cv2
import numpy as np

def sad(image1, image2):
    # Convert images to grayscale
    gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

    # Ensure both images have the same dimensions
    assert gray1.shape == gray2.shape, "Image dimensions do not match"

    # Calculate SAD
    sad_value = np.sum(np.abs(gray1 - gray2))

    return sad_value

# Load two images
image1 = cv2.imread('Lenna.png')
image2 = cv2.imread('Lenna.png')

# Calculate SAD between the two images
result_sad = sad(image1, image2)

# Display the result
print(f"Sum of Absolute Differences (SAD) between the two images: {result_sad}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

def create_gabor_filterbank(num_orientations, num_scales, size):
    filters = []
    for theta in range(num_orientations):
        theta = theta / num_orientations * np.pi
        for scale in range(num_scales):
            wavelength = size / np.sqrt(2)
            sigma = wavelength * 0.8
            sigma_x = sigma
            sigma_y = sigma / (2 * np.pi) * wavelength
            grid_x, grid_y = np.meshgrid(np.linspace(-1, 1, size),
                                         np.linspace(-1, 1, size))
            x_theta = grid_x * np.cos(theta) + grid_y * np.sin(theta)
            y_theta = -grid_x * np.sin(theta) + grid_y * np.cos(theta)
            filters.append(np.exp(-0.5 * (x_theta ** 2 / sigma_x ** 2 + y_theta ** 2 / sigma_y ** 2))
                           * np.cos(2 * np.pi / wavelength * x_theta))
    return filters

def apply_filterbank(image, filterbank):
    responses = [ndimage.convolve(image, filter, mode='nearest') for filter in filterbank]
    return responses

# Load an example image (replace 'example_image.jpg' with your image path)
image = plt.imread('Lenna.png')

# Create a Gabor filter bank
num_orientations = 8
num_scales = 5
filter_size = 15
filterbank = create_gabor_filterbank(num_orientations, num_scales, filter_size)

# Apply the filter bank to the image
responses = apply_filterbank(image, filterbank)

# Visualize the filter bank responses
plt.figure(figsize=(12, 12))
for i in range(len(responses)):
    plt.subplot(num_scales, num_orientations, i + 1)
    plt.imshow(np.abs(responses[i]), cmap='gray')
    plt.axis('off')
plt.show()
```
