```python
import cv2
import numpy as np
gambar = cv2.imread('./bola.jpg')

output = gambar.copy()
img = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)
# Cari iris (lingkaran)
circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 500, param1 = 50,
               param2 = 20, minRadius = 50, maxRadius = 150)
# jika ditemukan
if circles is not None:
   # cari koordinat x,y dan radius (r)
   circles = np.round(circles[0, :]).astype("int")
   print(circles)
   # cari terus dlm loop
   for (x, y, r) in circles:
      cv2.circle(output, (x, y), r, (0, 255, 0), 2)
# tampilkan gambar
cv2.imshow('Deteksi Lingkaran', output)
```

```python
import cv2 as cv
import numpy as np

# Nama file
filename = './bola.jpg'
# Load gambar traffic-sign.jpg
src = cv.imread(filename)

# resize ke 30% ukuran aslinya
scale_percent = 30
width = int(src.shape[1] * scale_percent / 100)
height = int(src.shape[0] * scale_percent / 100)
dim = (width, height)
resized = cv.resize(src, dim, interpolation = cv.INTER_AREA)

# ubah ke grayscale
gray = cv.cvtColor(resized, cv.COLOR_BGR2GRAY)
# beri filter median blur
gray = cv.medianBlur(gray, 5)

# aplikasikan hough circle
circles = cv.HoughCircles(gray, cv.HOUGH_GRADIENT, 0.5, 20,
                            param1=100, param2=30,
                            minRadius=10, maxRadius=30)
# Jika lingkaran ditemukan
if circles is not None:
    circles = np.uint16(np.around(circles))
    # Untuk setiap lingkaran, gambar lingkaran diatasnya
    for i in circles[0, :]:
        center = (i[0], i[1])
        # circle center
        cv.circle(resized, center, 1, (0, 100, 100), 3)
        # circle outline
        radius = i[2]
        cv.circle(resized, center, radius, (0, 234, 255), 3)

cv2.imshow('Deteksi Lingkaran', resized)
```

```python
import cv2
import numpy as np

# Baca gambar
img = cv2.imread('./bola2.jpg', cv2.IMREAD_GRAYSCALE)

# Lakukan pre-processing jika diperlukan (misalnya, ubah ukuran gambar, blurring, deteksi tepi, dll.)
# ...

# Deteksi lingkaran menggunakan Hough Transform
circles = cv2.HoughCircles(
    img,
    cv2.HOUGH_GRADIENT,
    dp=1,  # Resolusi ruang akumulator yang diinginkan (semakin kecil semakin akurat, tapi juga membutuhkan lebih banyak waktu)
    minDist=50,  # Jarak minimum antara pusat dua lingkaran yang dideteksi
    param1=50,  # Parameter deteksi tepi (semakin tinggi, semakin ketat)
    param2=30,  # Parameter ambang batas untuk memilih pusat lingkaran (semakin kecil, semakin ketat)
    minRadius=10,  # Radius minimum lingkaran yang akan dideteksi
    maxRadius=100  # Radius maksimum lingkaran yang akan dideteksi
)

# Jika lingkaran ditemukan, gambar mereka pada gambar asli
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # Gambar lingkaran di sekitar setiap deteksi
        cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)

# Tampilkan hasil
cv2.imshow('Deteksi Lingkaran', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
